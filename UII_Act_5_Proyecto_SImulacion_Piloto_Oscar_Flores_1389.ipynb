# Proyecto de Simulacion **Piloto**


```python

print("SISTEMA DE IA DEL AVIÓN ACTIVADO")
altitud = int(input("Ingrese la altitud actual en pies: ").replace(",", ""))
combustible = int(input("Ingrese el nivel de combustible (0-100): ").replace(",", ""))
clima = input("Ingrese el estado del clima (despejado, tormenta, niebla): ")

if combustible < 15:
    print("¡ALERTA CRÍTICA! Combustible por debajo del nivel de seguridad.")
    print("IA: Iniciando protocolos de aterrizaje de emergencia en la pista más cercana.")

elif clima == "tormenta" and altitud < 5000:
    print("IA: Detectada tormenta a baja altitud. Peligro de cizalladura.")
    print("Sugerencia de IA: Incremente la potencia y ascienda a 10,000 pies de inmediato.")

elif altitud > 35000:
    print("IA: Altitud superior al límite operativo normal.")
    print("IA: Ajustando mezcla de oxígeno y nivelando estabilizadores.")

elif clima == "niebla" or clima == "tormenta":
    print("IA: Visibilidad reducida detectada.")
    print("IA: Activando instrumentos de navegación por radio y piloto automático.")

elif 10000 <= altitud <= 30000 and combustible > 50:
    print("IA: Vuelo de crucero detectado en condiciones óptimas.")
    print("IA: Manteniendo velocidad constante. Todos los sistemas están normales.")

else:
    print("IA: Monitoreando parámetros de vuelo. Piloto, usted tiene el control manual.")

print("--- FIN DE LA EVALUACIÓN DE SISTEMAS ---")
print("Programa realizado por Oscar Flores NL = 22")
```
